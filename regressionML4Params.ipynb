```python
from pandas import read_csv, DataFrame
import matplotlib.pyplot as plt

df = read_csv('/gdrive/MyDrive/traitement_ordinaire/datarep/jeu_donnees2_MWQP.csv')
# plt.plot(df['PH'],df['taux'])
#plt.scatter(df['DO'],df['taux'])
# df=df[-30:]
df
```

```python
import numpy as np
from sklearn import linear_model
import pickle

from datetime import datetime, timedelta
import os

now = datetime.now()
directory=str('/gdrive/MyDrive/traitement_ordinaire/RegressionLineaire' +str( now.strftime(("%d_%b_%Y_%H_%M")) ) )
if not os.path.exists(directory):
    os.mkdir(directory)

#Create Linear Regression object
reg = linear_model.LinearRegression()

#Now let us call fit method to train the model using independent variables.
#And the value that needs to be predicted (Images_Analyzed)

model=reg.fit(df[['PH', 'EC', 'Temp','DO']], df.taux) #Indep variables, dep. variable to be predicted

pickle.dump(model, open(directory+"/regressor4params", 'wb'))
#Model is ready. Let us check the coefficients, stored as reg.coef_.
#These are a, b, and c from our equation. 
#Intercept is stored as reg.intercept_
# print(reg.coef_, reg.intercept_)

#All set to predict the number of images someone would analyze at a given time
# print(reg.predict([[6.67, 135.0, 26.0,52.67]]))
# print(reg.predict([[6.68, 572.0, 28.1,7.80]]))

# print(reg.predict([[7,125,24,7.5]])) # saison froide attendu 40.04 %. Obtenu 93.11
print( reg.predict([[6,220,27,3.5]])[0] ) #Saison chaude Compaor√©. Attendu 59.18% Obtenu = 1.31
```
